```python
import pandas as pd
```

```python
ColNames=['Rank', 'Player', 'Pos', 'Year Inducted', 'From', 'To', '1st Team All-Pro', 'Number of Pro-Bowls', 'Years as Primary Starter',
 'CarAV', 'Games Played', 'PassingCmp', 'PassingAtt', 'PassingYds', 'PassingTD', 'PassLng',
 'PassingInt', 'Sk', 'Yds.1', 'RushingAtt', 'RushingYds', 'RushingTD', 'RushLng', 'ReceivingRec',
 'ReceivingYds', 'ReceivingTD', 'RecLong']

url = "https://www.pro-football-reference.com/hof/"
df = pd.read_html(url)
hof_df = df[0]
hof_df.columns = ColNames
hof_df.drop(columns = ['PassLng', 'Sk', 'Yds.1', 'RushLng', 'RecLong'])
hof_df['Hall of Fame'] = 'Y'
hof_df
```

```python
hof_df.to_csv('hof.csv')
```
