```python
# import dependencies
import pandas as pd
import matplotlib as plt
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect
Base = automap_base()
```

```python
# get draft data
engine = create_engine('sqlite:///../data/fbdata.sqlite')
Base.metadata.create_all(engine)
Base.prepare(engine, reflect=True)
session = Session(engine)
draftDf = pd.read_sql_query("SELECT * FROM draftHistory WHERE Pos = 'QB'", con=engine)
print(draftDf.columns)
draftDf
```

```python
# CFB Stats
CFBStats = pd.read_sql_query("SELECT * FROM CFBStats", con=engine)
print(CFBStats.columns)
CFBStats
```

```python
# combine stats
combineStats = pd.read_sql_query("SELECT * FROM combineStats", con=engine)
print(combineStats.columns)
combineStats
```

```python
# HOF stats
hof = pd.read_sql_query("SELECT * FROM hof WHERE Pos = 'QB'", con=engine)
print(hof.columns)
hof.head(20)
```

```python
# ProBowl stats
probowl = pd.read_sql_query("SELECT * FROM probowl", con=engine)
print(probowl.columns)
probowl
```
